```python
import random
import matplotlib.pyplot as plt

def random_walk(steps):
    """Simulate a random walk for a given number of steps."""
    position = 0
    positions = [position]
    for _ in range(steps):
        step = random.choice([-1, 1])
        position += step
        positions.append(position)
    return positions

# Simulation parameters
num_steps = 1000
num_walks = 5

# Run multiple simulations
all_walks = [random_walk(num_steps) for _ in range(num_walks)]

# Plotting
plt.figure(figsize=(12, 6))
for i, walk in enumerate(all_walks):
    plt.plot(range(num_steps + 1), walk, label=f'Walk {i+1}')

plt.title('Random Walk Simulations')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Calculate and print some statistics
final_positions = [walk[-1] for walk in all_walks]
average_final_position = sum(final_positions) / len(final_positions)
max_distance = max(abs(pos) for pos in final_positions)

print(f"Average final position: {average_final_position}")
print(f"Maximum distance from start: {max_distance}")
```
